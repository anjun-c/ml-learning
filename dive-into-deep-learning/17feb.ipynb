```python
import torch
import torch.autograd as autograd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-2 * np.pi, 2 * np.pi, 100, requires_grad=True)
f = torch.sin(x)
x
```

```python
f
```

```python
f_prim = autograd.grad(f, x, torch.ones_like(f), create_graph=True)[0]
f_prim
```

```python
x_np = x.detach().numpy()
f_prim_np = f.detach().numpy()
grad_np = f_prim.detach().numpy()

plt.figure(figsize=(8, 5))
plt.plot(x_np, f_prim_np, label=r'$f(x) = \sin(x)$', linestyle='-', color='blue')
plt.plot(x_np, grad_np, label=r"$f'(x)$ (computed)", linestyle='--', color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title("Function and its Derivative using PyTorch Autograd")
plt.grid()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

nodes = ["x", "x^2", "log(x^2)", "sin(x)", "(log(x^2) * sin(x))", "x^(-1)", "f(x)"]
G.add_nodes_from(nodes)

edges = [
    ("x", "x^2"),
    ("x^2", "log(x^2)"),
    ("x", "sin(x)"),
    ("log(x^2)", "(log(x^2) * sin(x))"),
    ("sin(x)", "(log(x^2) * sin(x))"),
    ("x", "x^(-1)"),
    ("(log(x^2) * sin(x))", "f(x)"),
    ("x^(-1)", "f(x)"),
]

G.add_edges_from(edges)

plt.figure(figsize=(8, 6))
pos = {
    "x": (0, 2),
    "x^2": (-1, 1),
    "log(x^2)": (-2, 0),
    "sin(x)": (1, 1),
    "(log(x^2) * sin(x))": (-1, -1),
    "x^(-1)": (2, 0),
    "f(x)": (0, -2),
}

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='black', node_size=3000, font_size=10, font_weight='bold', arrows=True)
plt.title("Dependency Graph of f(x)")
plt.show()
```

6. Use the chain rule to compute the derivative 
<math xmlns="http://www.w3.org/1998/Math/MathML">
  <mfrac>
    <mrow>
      <mi>d</mi>
      <mi>f</mi>
    </mrow>
    <mrow>
      <mi>d</mi>
      <mi>x</mi>
    </mrow>
  </mfrac>
</math> of the aforementioned function, placing each term on the dependency graph that you constructed previously.

```python
G_deriv = nx.DiGraph()

nodes_deriv = [
    "x", 
    "x^2", "sin(x)", "x^(-1)",
    "log(x^2)",
    "(log(x^2) * sin(x))",
    "f(x)",
    "d(x^2)/dx = 2x", "d(log(x^2))/dx = 2/x", "d(sin(x))/dx = cos(x)",
    "d(log(x^2) * sin(x))/dx = (2/x * sin(x) + log(x^2) * cos(x))", 
    "d(x^(-1))/dx = -1/x^2",
    "df/dx"
]

G_deriv.add_nodes_from(nodes_deriv)

edges_deriv = [
    ("x", "x^2"), 
    ("x", "sin(x)"), 
    ("x", "x^(-1)"),
    
    ("x^2", "log(x^2)"),
    ("x^2", "d(x^2)/dx = 2x"),
    ("log(x^2)", "d(log(x^2))/dx = 2/x"),
    
    ("log(x^2)", "(log(x^2) * sin(x))"),
    ("sin(x)", "(log(x^2) * sin(x))"),
    ("sin(x)", "d(sin(x))/dx = cos(x)"),
    ("(log(x^2) * sin(x))", "d(log(x^2) * sin(x))/dx = (2/x * sin(x) + log(x^2) * cos(x))"),
    
    ("x^(-1)", "d(x^(-1))/dx = -1/x^2"),
    
    ("(log(x^2) * sin(x))", "f(x)"),
    ("x^(-1)", "f(x)"),
    
    ("f(x)", "df/dx"),
    ("d(log(x^2) * sin(x))/dx = (2/x * sin(x) + log(x^2) * cos(x))", "df/dx"),
    ("d(x^(-1))/dx = -1/x^2", "df/dx")
]

G_deriv.add_edges_from(edges_deriv)

pos_deriv = {
    "x": (0, 4),

    "x^2": (-4, 3),
    "sin(x)": (0, 3),
    "x^(-1)": (4, 3),
    
    "d(x^2)/dx = 2x": (-4, 2.6),
    "log(x^2)": (-4, 2),
    "d(log(x^2))/dx = 2/x": (-4, 1.6),
    "d(sin(x))/dx = cos(x)": (0, 2.6),
    "d(x^(-1))/dx = -1/x^2": (4, 2.6),
    
    "(log(x^2) * sin(x))": (-2, 1),
    "d(log(x^2) * sin(x))/dx = (2/x * sin(x) + log(x^2) * cos(x))": (-2, 0.5),
    
    "f(x)": (0, 0),
    "df/dx": (0, -1)
}

plt.figure(figsize=(10, 8))
nx.draw(
    G_deriv, pos_deriv, with_labels=True, node_color='lightblue', 
    edge_color='black', node_size=3000, font_size=9, font_weight='bold', arrows=True
)
plt.title("Clean Dependency Graph of f(x) with Derivatives")
plt.show()
```
